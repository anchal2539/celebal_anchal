# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        # Only allow safe characters (digits, operators, parentheses, spaces, decimal points)
        allowed_chars = set("0123456789+-*/(). ")
        if not set(expression) <= allowed_chars:
            return "Error in calculation"
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

````python
# 🛠️ TOOL 3 (BONUS): Simple Logger

def log_event(query: str, response: dict) -> None:
    """Log each query and the agent's response."""
    print(f"[LOG] Query: {query!r} -> Type: {response.get('type')}")
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION

import re

def agent(query: str):
    """
    Routes a user query to the correct tool based on intent, and returns
    a structured JSON-style response: {"type": ..., "result": ...}
    """
    try:
        if not query or not query.strip():
            response = {"type": "error", "result": "Empty query received."}
            log_event(query, response)
            return response

        query_lower = query.lower().strip()

        # --- Route 1: Calculator ---
        if "calculate" in query_lower:
            # Pull out just the mathematical expression from the query
            # e.g. "Calculate 20 + 5" -> "20 + 5"
            expr_match = re.search(r"calculate\s*(.*)", query_lower)
            expression = expr_match.group(1).strip() if expr_match else ""

            if not expression:
                response = {"type": "error", "result": "No expression found to calculate."}
            else:
                calc_result = calculator(expression)
                if calc_result == "Error in calculation":
                    response = {"type": "error", "result": calc_result}
                else:
                    response = {"type": "calculation", "result": calc_result}

        # --- Route 2: Keyword Extractor ---
        elif "keywords" in query_lower:
            # Strip common trigger phrases so only the actual text remains
            text = re.sub(r"(extract\s+)?keywords\s*(from)?\s*", "", query, flags=re.IGNORECASE).strip()

            if not text:
                response = {"type": "error", "result": "No text found to extract keywords from."}
            else:
                keywords = extract_keywords(text)
                response = {"type": "keywords", "result": keywords}

        # --- Route 3: General Response ---
        else:
            response = {
                "type": "general",
                "result": f"I don't have a specific tool for this, but you asked: '{query}'."
            }

    except Exception as e:
        response = {"type": "error", "result": f"Unexpected error: {str(e)}"}

    log_event(query, response)
    return response
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?",
    "Calculate 10 / 0",
    ""
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
